## if/else

```python
def make_sense_of_this_year(year: int):
    if year < 0:
        print("B.C")
        # year = -90 => halb praktika lisada funktsioon võrdluste sisse
        if year < -10000:
            print("Stone age...")
    elif year == 0:
        print("Happy birthday, Jesus!")
    elif year == 1978:
        print("Happy birthday, Mati!")
    else:
        print("A.D")

make_sense_of_this_year(-90)
make_sense_of_this_year(year=-90)
```

```python
def format_year(year: int) -> str:
    if year < 0:
        return (f"{abs(year)} B.C")
    else:
        return (f"{year} A.D")

print(format_year(-90))
print(format_year(year=2025))
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['target'] = iris_data.target
iris['species'] = iris['target'].map(dict(enumerate(iris_data.target_names)))
iris.pop('target')
iris = iris.rename(columns={
    "sepal length (cm)": "sepal_lenght",
    "sepal width (cm)": "sepal_width",
    "petal length (cm)": "petal_lenght",
    "petal width (cm)": "petal_width"
})
iris
```

```python
iris.to_csv("iris.csv", index=False)
```

- vaja leida kogu hind
- kui hind on üle 100, siis 20% alla
- kui hind on üle 50, siis 10% alla

sammud:
- leida allahindlusega hind igale hinnale
- kokku liita

```python
hinnad = [120, 45, 89, 200, 33, 75, 150, 10, 99, 101]

def get_discount(price: int) -> float: #funktsioon funktsiooni nimi(argumendi nimi: argumendi tüüp) -> tulemuse tüüp
    """
    seefunktsioon arvutab allahindlust
    """
    if price > 100:
        return price * 0.8
    if price > 50:
        return price * 0.9
    return price

def get_price(prices: list[int]) -> float: #loon funktsiooni allahinnatud ja mitte hindade summeerimiseks
    tulemus = 0
    for hind in prices:
        allahindlusega_hind = get_discount(hind)
        tulemus = tulemus + allahindlusega_hind
    return tulemus

def get_price_short(prices: list[int]) -> float:
    return sum([get_discount(hind) for hind in prices])

get_price_short(hinnad)

```
