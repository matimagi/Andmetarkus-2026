## Muutujad (variables)

```python
# muutujad on konteinerid erinevate väärtuste hoidmiseks
x = 5 # täisarvuline muutuja (integer variable)
y = "Hello, World!" # teksti rida/jada, muutuja (string variable)

print(x) # täisarvuline väljund (output integer)
print(y) # tekstiline väljund (output string)
```

```python
# muutuja tüüpi saab määrata muutuja loomisel
x = str(3) # tekst, x hakkab olema "3"
y = int(3) # täisarv, y hakkab olema 3
z = float(3) # ujukomanumber, z hakkab olema 3,0

print(x)
print(y)
print(z)

# kui soovid teada saada muutuja tüüpi, kasuta funktiooni type()

print(type(x))
print(type(y))
print(type(z))

# pythonis on muutujad tähekõrguse tundlikud

a = 5
A = "Mati"

# ...mis tähendab, et väike a ja suur A on erinevad muutujad

print(a)
print(A)
```

## Muutujate nimetused
Pythonis saab kasutada erineva pikkused muutujate nimesid, mis alluvad järgmistele reeglitele:
1. Muutuja nimetus peab algama tähega;
2. Muutuja nimetus ei tohi hakata numbriga;
    Mis tähendab, et "age2" on OK, aga "2age" ei ole OK !!!
3. Muutuja nimedesse sobivad kõik tähed, kõik numbrid ja alakriips kokku kirjutatuna;
    Mis tähendab et "my_var" on OK, aga "my var" või "my-var" ei ole OK !!!
4. Muutuja nimetused on tähesuuruse tundlikud;
    Mis tähendab, et need kolm muutujat on kõik erinevad: age = "15", Age = "30" ja AGE = "45" !!!
5. Muutujate nimetusteks ei saa olla märksõnad: if, else, while, for ja palju muid.

Kui kirjutada muutujate nimetusi mitmesõnalistena, siis on kolm varianti:
1. Läbiva suure tähega (Camel Case), kus iga sõna peale esimese algab suure tähega - **myVariableName**;
2. Läbiva suure tähega (Pascal Case), kus kõik sõnad kirjutatakse suure tähega - **MyVariableName**;
3. Alakriipsudega (Snake Case), kus kõik sõnad eraldatakse alakriipsuga - **my_variable_name**.

## Hulgimuutujad

```python
x, y, z = "Orange", "Apple", "Cherry" # muutujate ja väärtuste arv peab vastama, kui ei vasta,
# saame tulemuseks error'i

print(x)
print(y)
print(z)
```

```python
x = y = z = "Orange" # erinevatele muutujatele sama väärtuste lisamine

print(x)
print(y)
print(z)
```

```python
fruits = ["Orange", "Apple", "Cherry"] # unpacking a collection
x, y, z = fruits

print(x)
print(y)
print(z)
```

## Globaalsed muutujad (global variables)
Need on erilised muutujad, mis elavad väljaspool funktioone ja on ligipääsetevad nii funktsiooni seest, kui ka väljastpoolt. 

```python
x = "awesome"

def my_func():
    print ("Python is " + x)

my_func()
```

```python
x = "awesome"

def my_func():
    x = "fantastic" # kuna see muutuja on lokaalne, funktsiooni sisene, siis funktsioon kasutab seda
    print ("Python is " + x)

my_func()
```

```python
x = "awesome"

def my_func():
    x = "fantastic" # kuna see muutuja on lokaalne, funktsiooni sisene, siis funktsioon kasutab seda...
    print ("Python is " + x)

my_func()

print("Python is " + x) # ...aga väljaspool funktsiooni saame ikkagi kasutada globaalset muutujat
```

Nüüd, kui me soovime luua globaalse muutuja funktsiooni sees, me kasutame võtmesõna "global". See ütleb Pythonile, et me soovime, et meie muutuja oleks globaalne ja mitte lokaalne.

```python
def my_func():
    global x
    x = "fantastic"

my_func()

print("Python is " + x)
```

Esmalt me määrame x väärtuseks "awesome", tekitades sellega globaalse muutuja, seejärel muudame võtmesõnaga "global" globaalse muutuja funktsiooni enda sees "fantastic"-uks. Nüüd, kui me kuvame (läbi print() funktsiooni) x väljaspool funktsiooni, kuvatakse meile "fantastic".

```python
x = "awesome"

print("Python is " + x)

def my_func():
    global x
    x = "fantastic"

my_func()

print("Python is " + x)
```
