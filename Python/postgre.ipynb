```python
def numeric_year(year_str: str) -> int:
    """
    Convert a year in format "2025 A.D." or "2025 B.C." to numeric
    """
    # Ensure input has two components corresponding to year and era
    try:
        year_number_str, year_era_notation_str = year_str.split()
    except:
        raise ValueError(f"'{year_str}' is not a valid year number: exactly two components expected")

    # Ensure year number is numeric
    try:
        year_numeric = int(year_number_str)
    except:
        raise ValueError(f"'{year_str}' is not a valid year number: year not numeric")
    
    # Determin sign based on era identifier
    if year_era_notation_str == "A.D.":
        return year_numeric
    if year_era_notation_str == "B.C.":
        return -year_numeric
    raise ValueError(f"'{year_str}' does not seem to be a valid year number: unknown era")

print(numeric_year("2024 A.D.")) # 2024
print(numeric_year("1 B.C.")) # -1
print(numeric_year("12 A.D.")) # -1
# print(numeric_year("Kaksteist A.D.")) # 12
```
